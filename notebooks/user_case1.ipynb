### Search for ‘object domains’ [in sinhala and english generic representation] those have high correlation with the given ‘generic object domain’ 
- Metaphors that based on the given ‘object domain’
- ‘Singers’, & ‘Writers’  who songs have these ‘metaphors’
- ‘Song topic’ that highly correlates with the metaphors for given ‘object domain’


```python
from datetime import datetime
from elasticsearch import Elasticsearch
from collections import defaultdict
import json 
import pandas as pd 
```

```python
es = Elasticsearch(hosts="http://localhost:9200")
```

```python
index = 'song-corpus'
```

```python
resp = es.search(index="song-corpus", query={"match": {"Annotation.Subject-english": "lover"}})
hits = resp['hits']['hits']
hits
```

```python
# get all metaphors

all_annotation = []
all_data = {"Singer": [], "Musician": [], "Writers": [], "Topic": []}

for h in hits:
    h = h['_source']
    all_annotation.extend(h['Annotation'])
    try:
        all_data['Singer'].append((h['Singer'], h['title']))
    except:
        pass 

    try:
        all_data['Musician'].append((h['Music'], h['title']))
    except:
        pass 

    try:
        all_data['Writers'].append((h['Lyrics'], h['title']))
    except:
        pass 

    try:
        all_data['Topic'].append(h['topic'])
    except:
        pass 


all_annotation
```

```python
metaphors = [i['Metaphore'] for i in all_annotation]
object_sinhala = [i['Object-sinhala'] for i in all_annotation]
object_english = [i['Object-english'] for i in all_annotation]
subject_sinhala = [i['Subject-sinhala'] for i in all_annotation]
subject_english = [i['Subject-english'] for i in all_annotation]
```
